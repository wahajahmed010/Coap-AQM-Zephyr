```python
import serial
ser = serial.Serial('/dev/cu.usbmodem0010502514721', baudrate=115200, timeout=1)
while True:
    try:
        line = ser.readline().decode('utf-8').strip()
        if line:
            print(line)
    except Exception as e:
        print(f"Error reading from serial port: {e}")
        break
```
